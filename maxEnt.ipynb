## Shannon Entorpy:
$$ H(S) = - \sum_{x \in V} p(x) \log_2 p(x) $$


## Reward Function:

 $$ R(S) = \alpha \cdot H(S) + \beta \cdot E(S) - \gamma \cdot D_{KL}(S \parallel \text{English}) $$

### State 𝑆

- Current version of the modified English language:

- Vocabulary probabilities

- Grammar rules deviations (minor initially, larger later)

- Symbol encoding (tokenizations)

### KL-Divergence from English: 
- Regularizes divergence speed, preventing rapid incomprehensibility:

 $$ D_{KL}(S \parallel \text{English}) = \sum_{x \in V} p_S(x) \log \frac{p_S(x)}{p_{\text{English}}(x)} $$
